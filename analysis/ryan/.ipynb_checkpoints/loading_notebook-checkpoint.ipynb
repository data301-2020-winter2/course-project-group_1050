```python
%pip install seaborn
%pip install matplotlib
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
path_parent = os.path.dirname(os.getcwd())
os.chdir(path_parent)
print(os.getcwd())
from thierry.project_functions import load_and_process 
df = load_and_process('https://raw.githubusercontent.com/data301-2020-winter2/course-project-group_1050/main/data/raw/AirQualityUCI.csv')
df.describe()
```

```python
df[df['Temperature (C)'] > 40]
# this actually doesnt have a real purpose, but as you see above max temp recorded was 44c! (This is roadside temps!)
```

```python
corr = df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(250, 20, as_cmap=True))
# interestly all pollutants are positively correlated to eachother except for Tungston NO, however Tungston NO2 is positively correlated
# to pollutants so maybe Tungston NOx turns into Tungston NO2 or Tungston NO2 turns into Tungston NOX
```

```python
df.plot(kind='scatter', x='Tungsten Oxide (NOx)', y='Tungsten Oxide (NO2)')
# scatterplot does show negative correlation between Tungsten Oxide (NOx) & Tungsten Oxide (NO2)
```
